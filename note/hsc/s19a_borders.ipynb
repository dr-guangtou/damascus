### Get the polygon representations of HSC FDFC regions

#### HSC FDFC Healpix masks

```python
#s19a_fdfc = "/Users/song/Downloads/s19a_fdfc_hp_contarea_izy-gt-5_trimmed.fits"
s19a_fdfc = "/Users/song/astro5/massive/s19a/mask/fdfc/s19a_fdfc_hp_contarea_izy-gt-5_trimmed.fits"
s18a_fdfc = "/Users/song/astro5/massive/s18a/mask/s18a_fdfc_hp_contarea.fits"
```

#### S18A

```python
s18a_poly = hsc.get_fdfc_borders(
    s18a_fdfc, distance=1.0, alpha=0.1, use_boundary=False)

s18a_borders_poly = hsc.get_fdfc_borders(
    s18a_fdfc, distance=1.0, alpha=0.1, use_boundary=True)
```

```python
plt.figure(figsize=(14, 6))

for idx in np.arange(len(s18a_poly)):
    plt.plot(s18a_poly[idx + 1][:, 0], s18a_poly[idx + 1][:, 1], c='royalblue')
    plt.plot(s18a_borders_poly[idx + 1][:, 0], s18a_borders_poly[idx + 1][:, 1], c='orangered')
```

#### S19A 

```python
s19a_poly = hsc.get_fdfc_borders(
    s19a_fdfc, distance=1.0, alpha=0.1, use_boundary=False)

s19a_borders_poly = hsc.get_fdfc_borders(
    s19a_fdfc, distance=1.0, alpha=0.1, use_boundary=True)
```
